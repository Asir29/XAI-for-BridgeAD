### nuScenes : mini dataset download

##### nuScenes Dataset Folder Structure Explanation:

###### samples/ - Contains sensor data (LiDAR, camera, radar) for keyframes.
######             These are annotated frames captured at 2 Hz and are used
######             for tasks like detection and tracking.

###### sweeps/  - Contains unannotated intermediate sensor frames between keyframes.
######             These are useful for temporal data fusion or motion modeling.
######             LiDAR sweeps are recorded at ~20 Hz, providing richer point clouds.

###### maps/    - Contains HD map data including semantic layers (PNG images) and
#####             vectorized map information (JSON). These maps are used for
######             localization, planning, and map-aware perception tasks.

###### v1.0-*   - Contains metadata in JSON format (e.g., object annotations,
######             sensor calibration, scene splits). This is essential for linking
######             sensor data to annotations and understanding scene structure.
